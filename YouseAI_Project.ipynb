```python
def categorize_credit_limit(value):
    if value <= 500:
        return '0-500'
    elif 500 < value <= 1000:
        return '500-1000'
    elif 1000 < value <= 1500:
        return '1000-1500'
    elif 1500 < value <= 5000:
        return '1500-5000'
    elif 5000 < value <= 10000:
        return '5000-10000'
    else:
        return '10000 above'


data['Credit_Limit_Range'] = data['CREDIT_LIMIT'].apply(categorize_credit_limit)


print(data[['CREDIT_LIMIT', 'Credit_Limit_Range']].head())
```
